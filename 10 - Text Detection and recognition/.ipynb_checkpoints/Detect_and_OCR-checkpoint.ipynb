# Text detection with OpenCV (EAST) and OCR with Tesseract

The EAST model can detect text in natural scenes (rotated, vertical and horizontal text). On the other hand, the model does not recognize text, if we want to do it, we must think about other methods. As we know, Pytessaract offers a complete engine for text recognition, with several page segmentation modes. This notebook intends to build a two stages model to detect and recognize text in images.

The idea is simple, first we detect text with EAST model and post-process them, once we have the ROI, we OCR'ing with Pytessaract.

**Project Structure**

The project contains three folders. The first of them is the EAST module. This module determines which layers of the network we want to consider, in addition, it offers a pre-process function to convert the predictions into bounding box coordinates. The second folder contains the pre-trained model. The last folder is the dataset used in this example. The images are different scenarios and all of them contain text information.

## Importing Libraries


```python
# East module
from east import EAST_OUTPUT_LAYERS
from east import decode_predictions
# Standart modules
from imutils import paths
import pytesseract
import imutils
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

## Setting the dataset and the pre-trained model path

```python
dataset = 'dataset' #folder path
paths = list(paths.list_images(dataset)) # list of image paths
```

```python
model = 'model/frozen_east_text_detection.pb' # defining the model path
```

## Reading the test image

```python
image = cv.imread(paths[0])
```

**Grabbing the original height and width**

```python
(origH, origW) = image.shape[:2] #grabbing the image dimensions
```

**Defining the new image dimension and aspect ratio**

```python
#defining the new Width and height, calculating the aspect ratio
(newW, newH) = (320, 320)
rW = origW / float(newW)
rH = origH / float(newH)
```

## Defining the model

```python
net = cv.dnn.readNet(model)
```

**Preparing the image for predictions**

```python
blob = cv.dnn.blobFromImage(image, 1.0, (newW, newH),
                            (123.68, 116.78, 103.94), 
                            swapRB=True, crop=False) #preparing the image for predictions
```

```python
net.setInput(blob)# passing the transformed images
```

**making predictions**

Using the pre-trained model, we want to obtain the confidence score and the geometry informations to derive the text bounding boxes in the image.

```python
(scores, geometry) = net.forward(EAST_OUTPUT_LAYERS)
```

**Post processing**

The model gives us the confidences and informations about the possible bounding boxes, but to obtain the exact bounding boxes, we must post-process these informations. To perform it, we call the function `decode_prediction` that returns the confidences and the bounding boxes coordinates. You find the details of the function in the file `east.py`.

```python
(rects, confidence) = decode_predictions(scores, geometry)
```

**Selecting and OCR'ing the bounding boxes**

The function `decode_predictions` returns the possible bounding boxes and their confidences. We post-process them using the Non-maximum suppression with a confidence score of $0.5$. The problem when using Pytessaract to OCR these bounding boxes relies on the fact that they are shrunk bounding boxes, any recognition can return incorrect predictions. To tackle this problem, we pad the bounding boxes by a factor of $5\%$ of the total height and width.

```python
#Non-maxima suppression to obtain on single box
idxs = cv.dnn.NMSBoxesRotated(rects, confidence, 0.5, 0.4)

results = []#list to put the results
if len(idxs) > 0:
    
    for i in idxs.flatten():
        
        box = cv.boxPoints(rects[i])
        box[:, 0] *= rW # rescaling the width
        box[:, 1] *= rH # rescaling the height
        box = np.int0(box) # putting all bounding boxes coordinates together
        
        (x, y, w, h) = cv.boundingRect(box) #grabbing the bounding box coordinates without rotation
        
        # Padding the bounding box to have a better OCR
        # Delta step for x-y
        dX = int(w*0.05)
        dY = int(h*0.05)
        
        #Obtaining the new coordinates with the padding
        #startX = max(0, x - dX)
        startX = max(0, x - (dX * 2))
        #startY = max(0, y - dY)
        startY = max(0, y - (dY * 2))
        #endX = min(origW, x + w + (dX * 2))
        endX = min(origW, x + w + dX)
        #endY = min(origH, y + h + (dY * 2))
        endY = min(origH, y + h + dY)
        
        pROI = image[startY:endY, startX:endX] # new roi
        
        #OCR'ing the ROI
        option = '--psm 7 -c tessedit_char_blacklist=",Â®@.;%()|=/\}"' #blacklist
        text = pytesseract.image_to_string(pROI, config=option) #OCR'ing
        results.append((box, text)) #appending the bound box coordinates and the text
```

```python
results = sorted(results, key=lambda y: y[0][0][1]) # sorting the results botom to top
```

```python
print(x, y, w, h)
```

```python
results
```

**Visualizing the padded bounding box**

```python
plt.figure(figsize=(10,10))
plt.imshow(cv.cvtColor(pROI, cv.COLOR_BGR2RGB))
plt.show()
```

**Drawing the bounding box and the predicted text**

```python
output = image.copy()
for box, text in results:
    
    #Drawing the bounding boxes
    cv.polylines(output,[box], True, (0, 255, 0), 2)
    
    #Cleaning the text
    text = "".join([c if ord(c) < 128 else "" for c in text]).strip()
    print(text)
    #Grabbing the coordinates
    (x, y, w, h) = cv.boundingRect(box)
    cv.putText(output, text, (x, y + h + 20), cv.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 1)
    
```

**Visualizing the final result**

```python
plt.figure(figsize=(10,10))
plt.imshow(cv.cvtColor(output, cv.COLOR_BGR2RGB))
plt.show()
```

As we can see, for this simple image, the combination of EAST and Tessaract returns a satisfactory result. There is a commitment between the predicted bounding box and text. We are going to see in the above lines examples for more complex images.

## Detecting in multiples images

```python
count = 1
plt.figure(figsize=(20,20))
for i in paths:
    img = cv.imread(i)
    (origH, origW) = img.shape[:2] #grabbing the image dimensions
    #defining the new Width and height, calculating the aspect ratio
    (newW, newH) = (320, 320)
    rW = origW / float(newW)
    rH = origH / float(newH)
    
    net = cv.dnn.readNet(model)
    
    blob = cv.dnn.blobFromImage(img, 1.0, (newW, newH),
                            (123.68, 116.78, 103.94), 
                            swapRB=True, crop=False) #preparing the image for predictions
    
    net.setInput(blob)# passing the transformed image
    
    (scores, geometry) = net.forward(EAST_OUTPUT_LAYERS)
    
    (rects, confidence) = decode_predictions(scores, geometry)
    
    #Non-maxima suppression to obtain on single box
    idxs = cv.dnn.NMSBoxesRotated(rects, confidence, 0.5, 0.4)
    results = []#list to put the results
    if len(idxs) > 0:
    
        for i in idxs.flatten():

            box = cv.boxPoints(rects[i])
            box[:, 0] *= rW
            box[:, 1] *= rH
            box = np.int0(box)
            (x, y, w, h) = cv.boundingRect(box) #grabbing the bounding box coordinates without rotation
        
            # Padding the bounding box to have a better OCR
            # Delta step for x-y
            dX = int(w*0.05)
            dY = int(h*0.05)

            #Obtaining the new coordinates with the padding
            startX = max(0, x - (dX * 2))
            startY = max(0, y - (dY * 2))
            #endX = min(origW, x + w + (dX * 2))
            endX = min(origW, x + w + dX)
            #endY = min(origH, y + h + (dY * 2))
            endY = min(origH, y + h + dY)

            pROI = img[startY:endY, startX:endX]

            #OCR'ing the ROI
            option = '--psm 7 -c tessedit_char_blacklist=",Â®@.;%()|=/\}"' 
            text = pytesseract.image_to_string(pROI, config=option)
            results.append((box, text))

    results = sorted(results, key=lambda y: y[0][0][1])
    output = img.copy()
    for box, text in results:

        #Drawing the bounding boxes
        cv.polylines(output,[box], True, (0, 255, 0), 2)

        #Cleaning the text
        text = "".join([c if ord(c) < 128 else "" for c in text]).strip()

        #Grabbing the coordinates
        (x, y, w, h) = cv.boundingRect(box)
        cv.putText(output, text, (x, y + h + 20), cv.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 1)
    
    ax = plt.subplot(3,2,count)
    plt.imshow(cv.cvtColor(output, cv.COLOR_BGR2RGB))
    plt.axis('off')
    count += 1
plt.show()
```

## Conclusion

The EAST model is an efficient method to detect text in images. Merging the EAST model and Tesseract can build a good OCR engine. As we can see, just two images give perfect results. We must take into account that they are appropriate images. The image with the basketball player demands another page segmentation mode from Tesseract. For French health care cards, we must select the regions we want to OCR the image. Generally speaking, combining the two methods is not enough to solve all kinds of problems, but it is a starting point to build a complete OCR engine.
