# Extracting Passport Machine Readable Zone with OpenCV and Tesseract

Passports around the world have a common ground, the Machine Readable Zone (MRZ). The MRZ contains essentials personal information. Applications able to read this zone can reduce the waiting time in the immigration control. Furthermore, the technical used to extract the MRZ can be used for other task in computer vision. The code template presented here is based on the [PyImageSearch blog](https://www.pyimagesearch.com/2015/11/30/detecting-machine-readable-zones-in-passport-images/).

## Importing Libraries

```python
from imutils.contours import sort_contours
from imutils import paths
import numpy as np
import pytesseract as tsr
import imutils
import cv2 as cv
import sys
import matplotlib.pyplot as plt
```

## Loading and resizing the image

In this step we load the image and resize it conserving the aspect ratio.

```python
img = cv.imread('dataset/2.png')
img = imutils.resize(img, width=600)
(H,W) = img.shape[:2]
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

## Image pre-processing

To extract the MRZ, we perform some operations as:

- Convert into gray scale
- Image blurring
- Black Hat morphological operation
- Image gradient along the axis x
- Two Close morphological operation
- Image thresholding with Otsu method and binary operation
- Found and filter the contours
- Pad the bounding box
- OCR the image.

```python
#gray scale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(cv.cvtColor(gray, cv.COLOR_BGR2RGB))
```

```python
#defining kernels size
rectK = cv.getStructuringElement(cv.MORPH_RECT, (15,7))
sqK = cv.getStructuringElement(cv.MORPH_RECT, (21,21))
```

```python
#bluring
gray = cv.GaussianBlur(gray, (3,3), 0)
plt.imshow(cv.cvtColor(gray, cv.COLOR_BGR2RGB))
```

We apply the [Black Hat](https://docs.opencv.org/4.5.4/d4/d86/group__imgproc__filter.html#ga7be549266bad7b2e6a04db49827f9f32) as morphological operation. Black Hat considers the difference between the closed input image with the input image, it shows the dark region in the light background. We use the function [`cv.morphologyEx()`](https://docs.opencv.org/4.5.4/d4/d86/group__imgproc__filter.html#ga67493776e3ad1a3df63883829375201f), the passed arguments are image source, type of morphological operation and the kernel.

```python
bHat = cv.morphologyEx(gray, cv.MORPH_BLACKHAT, rectK)
plt.imshow(cv.cvtColor(bHat, cv.COLOR_BGR2RGB))
```

Now, we calculate the gradient along the axis $x$, using the function [`cv.Sobel()`](https://docs.opencv.org/4.5.4/d4/d86/group__imgproc__filter.html#gacea54f142e81b6758cb6f375ce782c8d). The arguments are the image source, the output image depth, order of the derivate along $x$ and $y$, the last is the kernel size. The reason to apply the derivate along the axis $x$ is to preserve the horizontal areas with vertical changes. We also consider the Scharr kernel passing `ksize = -1`.

```python
grad = cv.Sobel(bHat, ddepth = cv.CV_32F, dx = 1, dy = 0, ksize = -1)
grad = np.absolute(grad)
(minP, maxP) = (np.min(grad), np.max(grad))
grad = (grad - minP) / (maxP - minP)
grad = (grad*255).astype('uint8')
plt.imshow(cv.cvtColor(grad, cv.COLOR_BGR2RGB))
```

The two next steps are the Close morphological operation. The first operation has a rectangular kernel to reduce the space among the characters and then, the square kernel to reduce the gap among the lines.

```python
close = cv.morphologyEx(grad, cv.MORPH_CLOSE, rectK)
plt.imshow(cv.cvtColor(close, cv.COLOR_BGR2RGB))
```

```python
close = cv.morphologyEx(close, cv.MORPH_CLOSE, sqK)
plt.imshow(cv.cvtColor(close, cv.COLOR_BGR2RGB))
```

To finish the morphological operations, we apply the erode operation with [`cv.erode()`](https://docs.opencv.org/4.5.4/d4/d86/group__imgproc__filter.html#gaeb1e0c1033e3f6b891a25d0511362aeb). The arguments are image source, kernel size and number of iterations.

```python
close = cv.erode(close, None, iterations = 4)
plt.imshow(cv.cvtColor(close, cv.COLOR_BGR2RGB))
```

Image thresholding with Otsu method and binary operation.

```python
close = cv.threshold(close, 0, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)[1]
plt.imshow(cv.cvtColor(close, cv.COLOR_BGR2RGB))
```

Once we have the thresh image, we can find the contours and filter them. We consider the function [`cv.boundingRect()`](https://docs.opencv.org/4.5.4/d3/dc0/group__imgproc__shape.html#ga103fcbda2f540f3ef1c042d6a9b35ac7) to obtain the bounding box of each contour. We calculate the proportion of each bounding box, we select the bounding where the percentage of the width is greater than $0.8$ and less than $0.15$ for the percentage along the height.

```python
#finding contours and sort them
cnts = cv.findContours(close.copy(), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
cnts = sort_contours(cnts, method='bottom-to-top')[0] # botton-to-top, we consider it based on the fact that
# the mrz is always in the bottom of the image

#initializing the mzr
roi = None
output = img.copy()
for c in cnts:
    (x, y, w, h) = cv.boundingRect(c)
    percW = w / float(W)
    percH = h / float(H)
    
    #Filtering the bounding box
    if percW > 0.8 and percH < 0.15:
        cv.drawContours(output, [c], -1, (0, 0, 255), 3)
        roi = (x, y, w, h)
        break
plt.imshow(cv.cvtColor(output, cv.COLOR_BGR2RGB))
```

As we can see, the contour of the ROI does not include the entire text. We pad the MRZ by $3\%$.

```python
pX = int((x + w)*0.03)
pY = int((y + h)*0.03)
(x, y) = (x - pX, y - pY)
(w, h) = (w + (pX*2), h + (pY*2))
```

Creating the region of interest.

```python
roi = img[y : y + h, x : x + w].copy()
plt.imshow(cv.cvtColor(roi, cv.COLOR_BGR2RGB))
```

Now, we can OCR the image, passing the roi into the function `tsr.image_to_string(roi)`.

```python
mrz = tsr.image_to_string(roi)
mrz = mrz.replace(" ", "")
print(mrz)
```

As we can see, Tessaract obtained success in OCR'ing the image, the extracted text is the same in the image. We must keep in mind that, it won't always happen.

```python
mrz = mrz.replace('\n\x0c', '') # cleaning the text
```

Printing out the bounding box of the MRZ and the OCR'ed text.

```python
dy = 20 #gap between the lines
for i, line in enumerate(mrz.split('\n')):
    y_text = y - 30 + i*dy
    cv.putText(img, line, (x, y_text ), cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
cv.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
cv.imwrite('mrz_passport.jpg', img)
```

## OCR'ing multiples images

```python
# making the list of images
dataset = 'dataset' #dataset path
pathImages = list(paths.list_images(dataset))
```

**Single loop to OCR the dataset**

```python
count = 1
plt.figure(figsize=(20,20))
for i in pathImages:
    img = cv.imread(i)
    img = imutils.resize(img, width=600)

    #gray scale
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    (H,W) = gray.shape
    
    #defining kernels size
    rectK = cv.getStructuringElement(cv.MORPH_RECT, (15,7))
    sqK = cv.getStructuringElement(cv.MORPH_RECT, (21,21))

    #bluring and morphological operation
    gray = cv.GaussianBlur(gray, (3,3), 0)
    bHat = cv.morphologyEx(gray, cv.MORPH_BLACKHAT, rectK)
    grad = cv.Sobel(bHat, ddepth = cv.CV_32F, dx = 1, dy = 0, ksize = -1)
    grad = np.absolute(grad)
    (minP, maxP) = (np.min(grad), np.max(grad))
    grad = (grad - minP) / (maxP - minP)
    grad = (grad*255).astype('uint8')

    # finding lines
    close = cv.morphologyEx(grad, cv.MORPH_CLOSE, rectK)
    #applying again the close morphological operation, in this time with the squre kernel
    # to close the gap between the lines
    close = cv.morphologyEx(close, cv.MORPH_CLOSE, sqK)
    close = cv.erode(close, None, iterations = 4)
    close = cv.threshold(close, 0, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)[1]

    #finding contours and sort them
    cnts = cv.findContours(close.copy(), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    cnts = imutils.grab_contours(cnts)
    cnts = sort_contours(cnts, method='bottom-to-top')[0]

    #initializing the mzr

    roi = None
    for c in cnts:
        (x, y, w, h) = cv.boundingRect(c)
        percW = w / float(W)
        percH = h / float(H)

        #Filtering the bounding box
        if percW > 0.8 and percH < 0.15:
            roi = (x, y, w, h)
            break

    # Padding the bounding box
    pX = int((x + w)*0.03)
    pY = int((y + h)*0.03)
    (x, y) = (x - pX, y - pY)
    (w, h) = (w + (pX*2), h + (pY*2))
    
    #extracting the MRz from the image and cleaning up the text
    roi = img[y : y + h, x : x + w].copy()
    text = tsr.image_to_string(roi)
    text = text.replace(" ", "")
    text = text.replace('\n\x0c', '')
    # printing out the extraction
    dy = 20
    for i, line in enumerate(text.split('\n')):
        y_text = y - 30 + i*dy
        cv.putText(img, line, (x, y_text ), cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
    cv.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    text = None
    #plotting
    ax = plt.subplot(3,2,count)
    plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.axis('off')
    count += 1
plt.show()
```

This template for the passport image pre-processing works perfectly in all cases. About the OCR process, we do not have $100\%$ accurate results. Tesseract is not perfect, even if we have passed the MRZ, some characters are not well recognized. To improve it, we must think about deep learning models.
